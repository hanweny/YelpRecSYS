```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
users = pd.read_csv("../data/users2.csv")
restaurants = pd.read_csv('../data/restaurants2.csv')
reviews = pd.read_csv("../data/reviews_with_sentiment.csv")
proportion = pd.read_csv("../data/proportion.csv")
```

```python
proportion.columns = ['topic{}'.format(i) for i in range(10)]
reviews = pd.concat([reviews, proportion], axis=1)
reviews.head()
```

## Combine Topic Sentiment and Proportion to Restaurants

```python
business_positive = reviews.groupby('business_id')['positive'].mean().reset_index()
business_positive.head()
```

```python
business_topics = reviews.groupby('business_id')[proportion.columns].mean().reset_index()
business_topics.head()
```

```python
restaurants_combined = restaurants.merge(business_topics, left_on = 'business_id', right_on = 'business_id', how = 'left')
restaurants_combined = restaurants_combined.merge(business_positive, left_on = 'business_id', right_on = 'business_id', how = 'left')
```

```python
restaurants_combined.head()
```

```python
restaurants_combined.to_csv("../data/restaurants_with_topic_sentiment.csv", index = False)
```

## Combine topic proportion to users

```python
user_topics = reviews.groupby('user_id')[proportion.columns].mean().reset_index()
user_combined = users.merge(user_topics, left_on = 'user_id', right_on = 'user_id', how = 'left')
```

```python
user_combined.head()
```

```python
user_combined.to_csv("../data/users_with_topic_sentiment.csv", index = False)
```
